```python
df_rat = pd.read_csv('files/RAT/2020/1/1dez/REDS_RAT_20200101_20200110.csv', sep='|')

df_rat = df_rat.applymap(lambda x: x.strip() if type(x) == str else x)

df_rat_23 = df_rat[
        (df_rat['MUNICIPIO'] == 'ITAUNA') |
        (df_rat['MUNICIPIO'] == 'DIVINOPOLIS') |
        (df_rat['MUNICIPIO'] == 'ITATIAIUCU') |
        (df_rat['MUNICIPIO'] == 'CLAUDIO') |
        (df_rat['MUNICIPIO'] == 'CARMO DO CAJURU') |
        (df_rat['MUNICIPIO'] == 'ITAUNA')
      ]

df_rat_23.index = df_rat_23['RAT.NUM_ATIVIDADE']
df_rat_23.loc[:,'DTA_HRA_INICIO'] = df_rat_23['DTA_INICIO'] + " " + df_rat_23['HRA_INICIO']
df_rat_23.loc[:,'DTA_HRA_INICIO_DT'] = pd.to_datetime( df_rat_23['DTA_HRA_INICIO'], format='%d/%m/%Y %H:%M' )

df_rat_23.loc[:,'DTA_HRA_TERMINO'] = df_rat_23.loc[:,'DTA_TERMINO'] + " " + df_rat_23.loc[:,'HRA_TERMINO']
df_rat_23.loc[:,'DTA_HRA_TERMINO_DT'] = pd.to_datetime( df_rat_23.loc[:,'DTA_HRA_TERMINO'], format='%d/%m/%Y %H:%M' )

df_rat_23.loc[:,'TEMPO_DT'] = df_rat_23['DTA_HRA_TERMINO_DT'] - df_rat_23['DTA_HRA_INICIO_DT']

df_rat_23.loc[:,'TEMPO_INT'] = df_rat_23['TEMPO_DT'].dt.total_seconds() / 60

df_classificadores = pd.read_csv('tbl_validadores.csv')
```
