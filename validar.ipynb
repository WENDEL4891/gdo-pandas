```python
def class_setores(row):
    if row['Número REDS'] in df_classif['VALIDADOR'].values:
        return df_classif[df_classif['VALIDADOR']==row['Número REDS']]['SETOR'].values[0]
    elif row['Município'] != 'DIVINOPOLIS':
        return row['Município']
    elif row['Bairro'] in df_classif[df_classif['VALIDADOR_TIPO']=='BAIRRO']['VALIDADOR'].values:
        return df_classif[
            (df_classif['VALIDADOR_TIPO'] == 'BAIRRO') &
            (df_classif['VALIDADOR'] == row['Bairro'])
        ]['SETOR'].values[0]
    else:
        return 'other'

df_crimes['SETOR'] = df_crimes.apply(lambda row: class_setores(row), axis=1)

df_crimes[
    (df_crimes['Bairro'] != 'INVÁLIDO') &
    (df_crimes['Município'] == 'DIVINOPOLIS')
][['Bairro', 'SETOR']]
```
